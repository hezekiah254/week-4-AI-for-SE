```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

def test_login():
    driver = webdriver.Chrome()
    driver.get("https://example.com/login")  # Replace with actual login URL
    driver.maximize_window()

    # --- Test 1: Valid Credentials ---
    driver.find_element(By.ID, "username").send_keys("valid_user")
    driver.find_element(By.ID, "password").send_keys("valid_password")
    driver.find_element(By.ID, "loginBtn").click()
    time.sleep(2)

    if "dashboard" in driver.current_url:
        print("✅ Test 1 Passed: Valid credentials login successful")
    else:
        print("❌ Test 1 Failed: Valid login unsuccessful")

    # --- Test 2: Invalid Credentials ---
    driver.get("https://example.com/login")
    driver.find_element(By.ID, "username").send_keys("invalid_user")
    driver.find_element(By.ID, "password").send_keys("wrong_pass")
    driver.find_element(By.ID, "loginBtn").click()
    time.sleep(2)

    if "Invalid username or password" in driver.page_source:
        print("✅ Test 2 Passed: Invalid credentials properly rejected")
    else:
        print("❌ Test 2 Failed: Invalid credentials not handled correctly")

    driver.quit()

if __name__ == "__main__":
    test_login()
```
